Recordamos las ecuaciones en el Movimiento Rectilineo uniforme Acelerado, considerando que deseamos obtener la posición respecto al cambio de tiempo que existe y dado por el usuario. 

Sea $$
S=dessplazamiento
$$
$$
V_{o}=velocidad inicial
$$
$$
a=aceleración
$$
Entonces 

El desplazamiento está dado por la ecuación:

$$ 
S[i+1]=S[i]+V_{o}*Δt+1/2*a*Δt^2
$$

```julia
Δt = 0.125
t = (0:Δt:100) # Nuestro arreglo t, de tiempo, está en lapsos de Δt
```

```julia
Vo=96 #Asignamos un valor a la velocidad inicial
```

```julia
a=-9.8
```

```julia
s=zeros(length(t));
```

```julia
for i ∈ 1:length(t) # Para cada elemento que pertenece al intervalo de 1 hasta t
    if i != 1 #Si se cumple que el elemento tiene un valor diferente de 1 se opera
        s[i]=s[i-1]-Vo*Δt+1/2*a*Δt^2 #Siendo la posición igual al valor de esta en el elemento anterior más el producto de velocidad y tiempo
    end
end
```

```julia
using Plots
```

```julia
pyplot()
```

```julia
plot(t,s,w=2) 
title!("Movimiento rectílineo uniformemente acelerado") #Título principal
xaxis!("Tiempo (s)") #Nombres en abciss
yaxis!("Desplazamiento en (m)") #Nombre en ordenadas
```

```julia
variables=hcat(t,s)
```

```julia
writedlm("MRUA.csv", variables, '|')
```
